```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
x,y,z = np.loadtxt("SystemB/ExpectedOutput.csv",delimiter=',',unpack=True)
x1,y1,z1=np.loadtxt("SystemB/output.csv",delimiter=',',unpack=True)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.plot3D(x,y,z, color="green", linewidth=0.2)
ax.plot3D(x1,y1,z1, linewidth=0.2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
plt.plot(x,x1, label="dx")
plt.plot(y,y1, label="dy")
plt.plot(z,z1, label="dz")
```
